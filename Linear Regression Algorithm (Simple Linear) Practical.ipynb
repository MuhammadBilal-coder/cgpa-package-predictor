```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_excel(r"F:\UNIVERSITY\SEMESTER 6\MACHINE LEARNING\ML RELATED ALL CODE\CGPA4_vs_Package_500Students.xlsx")
```

```python
dataset.head(3)
```

```python
sns.scatterplot(x="CGPA_out_of_4", y ="Package_LPA", data=dataset)
plt.show()
```

```python
dataset.isnull().sum()
```

```python
x = dataset[["CGPA_out_of_4"]]
y = dataset["Package_LPA"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
# y = 2.98386773x + 1.2789260030885643
```

```python
lr.score(x_test, y_test)*100
```

```python
lr.predict([[2.16 ]])
```

```python
 2.98386773 *2.16  + 1.2789260030885643
```

```python
y_prd = lr.predict(x)
```

```python
sns.scatterplot(x="CGPA_out_of_4", y ="Package_LPA", data=dataset)
plt.plot(dataset["CGPA_out_of_4"], y_prd, c= "red")
plt.legend(["original Data", "predict line"])
plt.savefig('graph.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import joblib

joblib.dump(lr, "model.pkl")
print("saved model.pkl")
```
